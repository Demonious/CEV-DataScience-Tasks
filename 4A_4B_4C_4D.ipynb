TASK-4B: Please provide the visual illustrations of the dataset that will demonstrate the insightful distributions with respect to the target column (CDRGLOB: the level of the Alzheimer Disease -AD) 

```python
import numpy as np
import matplotlib.pyplot as plt

x = data ["EDUC"]
y = data["SMOKYRS"]
labels = data["CDRGLOB"]
plt.scatter(x, y, s=20, cmap = "viridis", c=labels, alpha=0.5)
plt.xlabel("Education Years")
plt.ylabel("Smoking Years")
plt.title("Edu-Smk Scatter")
plt.colorbar()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = data ["NACCBMI"]
y = data["HIPPOVOL"]
labels = data["CDRGLOB"]
plt.scatter(x, y, s=20, cmap = "viridis", c=labels, alpha=0.5)
plt.xlabel("NACCBMI")
plt.ylabel("Hippocampus Volume")
plt.title("BMI-HippoVol Scatter")
plt.colorbar()
plt.show()
```

TASK-4D: Please provide several pivot tables for insightful illustrations both via tables and figures.

```python
hippovol = pd.qcut(data["HIPPOVOL"], 4)
data.pivot_table("NACCAGEB", index="CDRGLOB", columns=hippovol)
```

```python
data.pivot_table("HIPPOVOL", index=["CDRGLOB","NACCAGEB"], columns=hippovol).plot(kind="bar")
plt.legend() 
```

```python
hippovol = pd.qcut(data["HIPPOVOL"], 2)
data.pivot_table("HIPPOVOL", index=["CDRGLOB","NACCAGEB"], columns=["INDEPEND", hippovol])
```

```python
hippovolv2 = pd.qcut(data["HIPPOVOL"], 4)
data.pivot_table("SEX", index="CDRGLOB", columns=hippovolv2)
```
